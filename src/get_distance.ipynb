```python
import pandas as pd
import requests
```

```python
df_locations = pd.read_csv('./final_data/ntu_station_locations_all.csv')
df_locations.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], inplace = True)
df_locations
```

- Get the locations of the selected points

```python
candidate = ["國立臺灣大學圖書館", "國立臺灣大學管理學院", "台灣大學行政大樓", "國立臺灣大學博雅教學館", "國立臺灣大學醉月湖", "國立臺灣大學綜合體育館", "國立臺灣大學 計算機及資訊網路中心", "國立臺灣大學工學院綜合大樓", "第一學生活動中心", "國立臺灣大學綜合教學館", "臺灣大學電機資訊學院", "國立臺灣大學社會科學院", "台大小福廣場", "美德耐 台科大活動中心第一餐廳", "麗陽商場 台科大第三學生餐廳"] 
lat = []
lng = []
api_key = ""

for name in candidate:
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input="+ name + "&inputtype=textquery&fields=geometry&key="+ api_key
    payload= {}
    headers = {}
    response = requests.request("GET", url, headers=headers, data=payload)
    res_json = response.json()
    lat.append(res_json["candidates"][0]["geometry"]["location"]['lat'])
    lng.append(res_json["candidates"][0]["geometry"]["location"]['lng'])

candidate_dict = {"Name":candidate, "Latitude":lat, "Longitude":lng}
df_candidate = pd.DataFrame(candidate_dict)
df_candidate
```

```python
df_candidate.to_csv("candidates.csv")
```

- Get the distance matrix

```python
res = {"From":[], "To":[], "Distance (meters)":[], "Duration (seconds)":[], "Type":[]}
station_names = list(df_locations["Name"])
candidate_names = list(df_candidate["Name"])
print(station_names, candidate_names)
```

```python
for station in station_names:
    station_lat = float(df_locations[df_locations["Name"] == station]["Latitude"])
    station_lng = float(df_locations[df_locations["Name"] == station]["Longitude"])
    # 15 Candidates 
    for c, can in enumerate(candidate_names):
        can_lat = float(df_candidate[df_candidate["Name"] == can]["Latitude"])
        can_lng = float(df_candidate[df_candidate["Name"] == can]["Longitude"])
        url = "https://maps.googleapis.com/maps/api/distancematrix/json?origins="+ str(station_lat) + "%2C" + str(station_lng) + "&destinations="+ str(can_lat) + "%2C" + str(can_lng)  +"&mode=bicycling&region=tw&key=" + api_key
        payload={}
        headers= {}
        res_2 = requests.request("GET", url, headers=headers, data=payload).json()
        distance  = res_2["rows"][0]["elements"][0]["distance"]["value"]
        duration  = res_2["rows"][0]["elements"][0]["duration"]["value"]
        res["From"].append(station)
        res["To"].append(can)
        res["Distance (meters)"].append(distance)
        res["Duration (seconds)"].append(duration)
        res["Type"].append("Candidate"+str(c))
    # Nearest MRT (16)
    mrt_name = (df_locations[df_locations["Name"] == station]["MRT Station Name"])
    mrt_lat = float(df_locations[df_locations["Name"] == station]["Latitude_m"])
    mrt_lng = float(df_locations[df_locations["Name"] == station]["Longitude_m"])
    url = "https://maps.googleapis.com/maps/api/distancematrix/json?origins="+ str(station_lat) + "%2C" + str(station_lng) + "&destinations="+ str(mrt_lat) + "%2C" + str(mrt_lng)  +"&mode=bicycling&region=tw&key=" + api_key
    payload={}
    headers = {}
    res_2 = requests.request("GET", url, headers=headers, data=payload).json()
    distance  = res_2["rows"][0]["elements"][0]["distance"]["value"]
    duration  = res_2["rows"][0]["elements"][0]["duration"]["value"]
    res["From"].append(station)
    res["To"].append(mrt_name)
    res["Distance (meters)"].append(distance)
    res["Duration (seconds)"].append(duration)
    res["Type"].append("MRT")
    # Nearest Dorm (17)
    dorm_name = (df_locations[df_locations["Name"] == station]["Dorm Name"]).to_string()
    dorm_lat = float(df_locations[df_locations["Name"] == station]["Latitude_Dorm"])
    dorm_lng = float(df_locations[df_locations["Name"] == station]["Longitude_Dorm"])
    url = "https://maps.googleapis.com/maps/api/distancematrix/json?origins="+ str(station_lat) + "%2C" + str(station_lng) + "&destinations="+ str(dorm_lat) + "%2C" + str(dorm_lng)  +"&mode=bicycling&region=tw&key=" + api_key
    payload={}
    headers = {}
    res_2 = requests.request("GET", url, headers=headers, data=payload).json()
    mrt_name = (df_locations[df_locations["Name"] == station]["MRT Station Name"]).to_string()
    distance  = res_2["rows"][0]["elements"][0]["distance"]["value"]
    duration  = res_2["rows"][0]["elements"][0]["duration"]["value"]
    res["From"].append(station)
    res["To"].append(dorm_name)
    res["Distance (meters)"].append(distance)
    res["Duration (seconds)"].append(duration)
    res["Type"].append("Dorm")

df_res = pd.DataFrame(res)
```

```python
df_res.rename(columns={"Distance (meters)":"dist"}, inplace=True)
df_res.drop(columns=["To", "Duration (seconds)"], inplace=True)
df_res
```

```python
df_res.to_csv('./final_data/distance_matrix.csv',index=False)
```
